# Your First Agent

Learn how to create and use your first Victor AI agent.

## Quick Start with Ollama (Recommended)

We'll use **Ollama** for local AI - it's free, fast, and runs locally.

### Setup Ollama

```bash
# Install Ollama
brew install ollama  # macOS
# OR
curl -fsSL https://ollama.com/install.sh | sh

# Start Ollama (runs in background)
ollama serve &

# Pull a model with tool support
ollama pull llama3.1:8b
```

### Verify Ollama is Running

```bash
# Check if Ollama is running
curl http://localhost:11434/api/tags

# List available models
ollama list
```

### Custom Ollama Endpoint

If Ollama is hosted on a different server, set the `OLLAMA_HOST` environment variable:

```bash
# For remote Ollama server
export OLLAMA_HOST="http://your-server:11434"

# Or use a custom port
export OLLAMA_HOST="http://localhost:8080"
```

Then create agents normally - Victor will use the custom endpoint:

```python
agent = await Agent.create(
    provider="ollama",
    model="llama3.1:8b"
)
```

## Basic Agent with Ollama

Create a simple agent using Ollama:

````python
from victor.framework import Agent

# Create agent with Ollama (local model)
# Using llama3.1:8b which has official tool support
agent = await Agent.create(
    provider="ollama",
    model="llama3.1:8b"  # 8B parameters, good tool support
)

# Run a query
result = await agent.run("What is 2 + 2? Reply with just the number.")
print(f"Response: {result.content}")
````

## Streaming Responses

For real-time feedback, use streaming:

````python
from victor.framework import Agent

agent = await Agent.create(
    provider="ollama",
    model="llama3.1:8b"
)

# Stream response as it arrives
async for event in agent.stream("Count to 3. Reply with just the numbers."):
    if event.type == "content":
        print(event.content, end="", flush=True)
    elif event.type == "tool_call":
        print(f"\n[Tool: {event.tool_name}]")
    elif event.type == "thinking":
        print("\n[Thinking...]")
````

## Multi-turn Conversations

Maintain context across multiple messages:

````python
from victor.framework import Agent

agent = await Agent.create(
    provider="ollama",
    model="llama3.1:8b"
)

# First message
response1 = await agent.run("My favorite color is blue. Remember this.")
print(f"Agent: {response1.content}")

# Second message - context depends on session management
# For simple queries, use run() with full context
response2 = await agent.run("What's my favorite color? It's blue.")
print(f"Agent: {response2.content}")
````

## Using Different Ollama Models

Ollama supports many models. Here are some popular options with tool support:

| Model | Size | Tool Support | Use Case |
|------|------|--------------|----------|
| llama3.1:8b | 4.9 GB | ✓ Official | General purpose (recommended) |
| llama3.2:3b | ~2GB | ✓ Official | Fast, resource-constrained |
| mistral:7b | ~4.1GB | ✓ Official | Fast, good quality |
| qwen2.5-coder:7b | 4.7GB | ⚠ Limited | Coding specialized |

### Pulling Models

```bash
# Pull the recommended model
ollama pull llama3.1:8b

# Pull a smaller model for faster responses
ollama pull llama3.2:3b

# Pull a coding-specialized model
ollama pull qwen2.5-coder:7b
```

### Verifying Model Availability

Before using a model, verify it's available:

```bash
ollama list | grep llama3.1
```

Or check programmatically:

```python
import subprocess

def check_model(model_name: str) -> bool:
    result = subprocess.run(
        ["ollama", "list"],
        capture_output=True,
        text=True
    )
    return model_name in result.stdout

if check_model("llama3.1:8b"):
    print("✓ Model available")
else:
    print("✗ Model not found, run: ollama pull llama3.1:8b")
```

## Troubleshooting Ollama

### Ollama not responding?

```bash
# Check if Ollama is running
ps aux | grep ollama

# Check if API is accessible
curl http://localhost:11434/api/tags

# Restart Ollama
killall ollama
ollama serve &
```

### Model not found?

```bash
# List available models
ollama list

# Pull the model
ollama pull qwen2.5-coder:7b
```

### Slow responses?

- Use a smaller model (e.g., `llama3.2:3b` instead of `qwen2.5-coder:7b`)
- Make sure no other apps are using Ollama heavily
- Check available RAM: `htop` or `Activity Monitor`

### Wrong responses or errors?

- Try pulling the model again: `ollama pull qwen2.5-coder:7b`
- Check Ollama logs: `~/.ollama/logs/server.log`
- Try a different model: `ollama pull llama3.2:3b`
