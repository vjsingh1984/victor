# Your First Agent

Learn how to create and use your first Victor AI agent.

## Prerequisites

You need an LLM provider API key or a local model.

### Option 1: Local Model (Ollama)

```bash
# Install Ollama
brew install ollama  # macOS
# OR
curl -fsSL https://ollama.com/install.sh | sh

# Start Ollama and pull a model
ollama serve &
ollama pull qwen2.5-coder:7b
```

### Option 2: Cloud Provider

```python
import os
os.environ['ANTHROPIC_API_KEY'] = 'your-key-here'
```

## Basic Agent

Create a simple agent and run a query:

````python
from victor.framework import Agent

# Create agent with default settings
agent = await Agent.create()

# Run a query
result = await agent.run("Explain quantum computing in simple terms.")
print(result.content)
````

## Agent with Specific Provider

Choose a specific LLM provider:

````python
# Use Ollama (local model)
agent = await Agent.create(
    provider="ollama",
    model="qwen2.5-coder:7b"
)

result = await agent.run("What is Python?")
print(result.content)
````

## Streaming Responses

For real-time feedback, use streaming:

````python
agent = await Agent.create()

# Stream response as it arrives
async for event in agent.stream("Tell me a short story"):
    if event.type == "content":
        print(event.content, end="", flush=True)
    elif event.type == "tool_call":
        print(f"\n[Tool: {event.tool_name}]")
    elif event.type == "thinking":
        print("\n[Thinking...]")
````

## Multi-turn Conversations

Maintain context across multiple messages:

````python
agent = await Agent.create()

# First message
response1 = await agent.chat("My favorite color is blue")
print(f"Agent: {response1.content}")

# Agent remembers context
response2 = await agent.chat("What's my favorite color?")
print(f"Agent: {response2.content}")
````
