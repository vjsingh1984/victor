# Multi-Agent Teams

Coordinate multiple agents.

```python
from victor.framework import Agent
from victor.teams import AgentTeam, TeamMemberSpec, TeamFormation

# Create agents
researcher = await Agent.create()
coder = await Agent.create()

# Create team
team = AgentTeam(
    name="dev-team",
    formation=TeamFormation.SEQUENTIAL,
    members=[
        TeamMemberSpec(agent=researcher, name="researcher"),
        TeamMemberSpec(agent=coder, name="coder"),
    ],
)

# Run team
agent = await Agent.create()
result = await agent.run_team(team, "Implement a REST API")
print(result.final_answer)
```
