# Multi-Agent Teams

Coordinate multiple agents working together.

## Prerequisites

This notebook requires an LLM provider. The examples below use **Ollama** running locally with the `qwen2.5-coder:7b` model.

### Ollama Setup

```bash
# Install Ollama (macOS/Linux)
brew install ollama

# Start Ollama server
ollama serve

# Pull the recommended model (7B parameters, ~4.5GB)
ollama pull qwen2.5-coder:7b
```

### Alternative Ollama Models

| Model | Size | Best For | Command |
|-------|------|----------|---------|
| `qwen2.5-coder:7b` | ~4.5GB | Coding, general | `ollama pull qwen2.5-coder:7b` |
| `llama3.2:3b` | ~2GB | Fast responses | `ollama pull llama3.2:3b` |
| `mistral:7b` | ~4.2GB | General tasks | `ollama pull mistral:7b` |

### Troubleshooting

If Ollama is not running:
```bash
# Check if Ollama is running
ps aux | grep ollama

# Start Ollama server
ollama serve &
```

To verify Ollama is working:
```bash
ollama list
ollama run qwen2.5-coder:7b "Hello"
```

````python
from victor.framework import Agent, TeamFormation, TeamMemberSpec

# Create a multi-agent team using Ollama
# Using llama3.1:8b which has official tool support
team = await Agent.create_team(
    name="api-implementation",
    goal="Implement a REST API for user management",
    provider="ollama",
    model="llama3.1:8b",
    formation=TeamFormation.SEQUENTIAL,
    members=[
        TeamMemberSpec(role="researcher", goal="Find best practices for REST APIs"),
        TeamMemberSpec(role="architect", goal="Design the API structure"),
        TeamMemberSpec(role="coder", goal="Write the implementation code"),
    ],
)

# Run the team
result = await team.run()
print(result.final_output)
````
