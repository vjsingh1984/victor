# Production Tips

Best practices for production deployment.

## Error Handling

```python
from victor.framework import Agent
from victor.framework.exceptions import ToolError, ProviderError

# Create agent with Ollama
agent = await Agent.create(
    provider="ollama",
    model="qwen2.5-coder:7b",
)

try:
    result = await agent.run("Process data")
except ToolError as e:
    print(f"Tool failed: {e.tool_name}")
except ProviderError as e:
    print(f"Provider error: {e.message}")
```

## Observability

```python
# Enable metrics with Ollama
agent = await Agent.create(
    provider="ollama",
    model="qwen2.5-coder:7b",
    enable_observability=True
)
```

## Session Management

```python
from victor.framework import Agent

# Create agent with Ollama
agent = await Agent.create(
    provider="ollama",
    model="qwen2.5-coder:7b",
)

# Use session for context persistence
result = await agent.run(
    "Continue our work",
    session_id="user-123"
)
```

## Rate Limiting

```python
from victor.framework import Agent, RateLimiter

# Create agent with rate limiting for production
agent = await Agent.create(
    provider="ollama",
    model="qwen2.5-coder:7b",
    rate_limiter=RateLimiter(max_requests=100, window=60)
)
```

## Provider Fallback

```python
from victor.framework import Agent

# Configure fallback providers for resilience
agent = await Agent.create(
    provider="ollama",
    model="qwen2.5-coder:7b",
    fallback_providers=[
        ("ollama", "llama3.2:3b"),  # Fallback to smaller model
    ]
)
```
